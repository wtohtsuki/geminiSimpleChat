# Configurações Iniciais

```python
import numpy as np  # processamento de matrizes multi-dimensionais
import pandas as pd  # manipulação e análise de dados
import google.generativeai as genai
from google.colab import userdata

GOOGLE_APY_KEY = userdata.get('secret_01')
genai.configure(api_key=GOOGLE_APY_KEY)
```

# Modelos / Métodos disponíveis

```python
for m in genai.list_models():
  if 'generateContent' in m.supported_generation_methods:
    print(f'name: {m.name}')
```

# Configurações do Modelo

```python
model_name = 'gemini-1.0-pro'

generation_config = {
    'candidate_count': 1,
    'temperature': 0.7  #
}

safety_settings = {  # All types of configuration
    'harassment': 'BLOCK_NONE',
    'hate': 'BLOCK_LOW_AND_ABOVE',
    'sexual': 'BLOCK_MEDIUM_AND_ABOVE',
    'dangerous': 'BLOCK_ONLY_HIGH',
}

model = genai.GenerativeModel(model_name=model_name,
                              generation_config=generation_config,
                              safety_settings=safety_settings)

print(model.model_name)
```

```python
response = model.generate_content('What are the most commom genres in anime?')

try:
  print(response.text)
except Exception as error:
  print('Error occurred!', error)  # included due to error on 1st run, haven't occcurred after that.
  print(response)
```

```python
embedding_model = 'models/embedding-001'

def embed_fn(title, content):
  return genai.embed_content(model=embedding_model,
                             title=title,
                             content=content,
                             task_type="RETRIEVAL_DOCUMENT")["embedding"]

df['embeddings'] = df.apply(lambda x: embed_fn(x['title'], x['content']), axis=1)
df
```

```python
def get_content_fn(query, data_frame, embedding_model):
  embedding_result = genai.embed_content(model=embedding_model,
                                   title=None,
                                   content=query,
                                   task_type="RETRIEVAL_DOCUMENT")["embedding"]

  dot_product = np.dot(np.stack(data_frame["embeddings"]), embedding_result)

  index = np.argmax(dot_product)

  return data_frame.iloc[index]['content']


query = 'Chibi Marukochan would be considered what type of anime?'

response = get_content_fn(query, df, embedding_model)

print(response)
```
